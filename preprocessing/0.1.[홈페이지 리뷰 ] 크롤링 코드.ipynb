```python
import re
import os
import sys
import math
import time
import random
import requests
sys.path.append('../')
import datetime
import pandas as pd
from tqdm import tqdm
from bs4 import BeautifulSoup
from selenium import webdriver
from urllib.request import urlopen
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException as NSEE
```

```python
driver = webdriver.Chrome()  # chrome driver 입력
driver.implicitly_wait(10)

url = 'https://www.kurly.com/collections/beauty-best?site=beauty&page=1&filters=category%3A363&per_page=96&sorted_type=4'

driver.get(url)
driver.implicitly_wait(10)
time.sleep(5)


#카테고리 따기
category = []
text_path = '/html/body/div/div[3]/div[1]/div[2]/div[2]/div[2]/div/span'
category = driver.find_element(By.XPATH,text_path).text
#print(category)
```

```python
reviews = []
likes = []
titles = []  # 'titles'라는 이름으로 변경하고 반복문 바깥에서 한 번만 초기화
dates = []    

# 상품 클릭
for i in range(1, 5):
    try:
        driver.find_element(By.XPATH, '/html/body/div/div[3]/div[1]/div/div[2]/div[3]/a[' + str(i) + ']/div[1]/div/span/img').click()
    except:
        print("상품 갯수 부족")

    # 상품명 따기
    text_path = '/html/body/div/div/div[3]/div[2]/main/section/div[2]/h1'
    title = driver.find_element(By.XPATH, text_path).text
    titles.append(title)  # 'append'를 사용하여 'titles' 리스트에 추가
    #print(title)

    # 리뷰로 이동
    try:
        driver.find_element(By.XPATH, '/html/body/div/div/div[3]/div[2]/nav/ul/li[3]/a/span[1]').click()
    except:
        print("리뷰 갯수 오류 발생")

    time.sleep(1)

    for j in range(4, 14):
        review_text_path = '/html/body/div/div/div[3]/div[2]/div/div[3]/section/div[2]/div[' + str(j) + ']/article/div/p'
        review = driver.find_element(By.XPATH, review_text_path).text
        reviews.append(review)  # 'append'를 사용하여 'reviews' 리스트에 추가
        #print(review)

        like_text_path = '/html/body/div/div/div[3]/div[2]/div/div[3]/section/div[2]/div[' + str(j) + ']/article/div/footer/button/span[2]'
        like = driver.find_element(By.XPATH, like_text_path).text
        likes.append(like)  # 'append'를 사용하여 'likes' 리스트에 추가
        #print(like)
        
        date_text_path = '/html/body/div/div/div[3]/div[2]/div/div[3]/section/div[2]/div[' + str(j) + ']/article/div/footer/div/span'
        date = driver.find_element(By.XPATH, date_text_path).text
        dates.append(date)  # 'append'를 사용하여 'date' 리스트에 추가
        #print(date)

    # 다시 이동 페이지로 이동
    driver.get(url)
    driver.implicitly_wait(10)
    time.sleep(5)
```

```python
import pandas as pd
import re

data = []
for i in range(4):  # 각 제품에 대해 반복
    for j in range(10):  # 각 제품의 리뷰에 대해 반복
        index = i * 10 + j
        data.append({
            'Category': category,
            'Product Name': titles[i],
            'Review': reviews[index],
            'Likes': likes[index],
            'Dates': dates[index]
        })

df = pd.DataFrame(data)

# CSV 파일로 저장
df.to_csv('reviews.csv', index=False)

# 지정된 경로에 CSV 파일로 저장
save_path = r'C:/Users/kgw57/PycharmProjects/data_contest/review_beauty/reviews_beauty_남성화장품.csv'
df.to_csv(save_path, index=False)
```
